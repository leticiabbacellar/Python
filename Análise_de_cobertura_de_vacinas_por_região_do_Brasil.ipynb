```python
import pandas as pd
import plotly.express as px
```

```python
df= pd.read_csv("/content/cpnibr17476108949.csv", sep=';', encoding='cp1252')
df
```

```python
for col in df.columns[1:]:
  df[col]= df[col].str.replace(",", ".").astype(float)

df
```

```python
df = df[df["Ano"] != "Total"]
```

```python
df_at= df.melt(
    id_vars= ["Ano"],
    value_vars= ["1 Região Norte",
                 "2 Região Nordeste",
                 "3 Região Sudeste",
                 "4 Região Sul",
                 "5 Região Centro-Oeste"],
    var_name= "Regiao",
    value_name= "Cobertura"
)
coordenadas = pd.DataFrame({
    "Regiao": [
        "1 Região Norte",
        "2 Região Nordeste",
        "3 Região Sudeste",
        "4 Região Sul",
        "5 Região Centro-Oeste"
    ],
    "Latitude": [-3.4653, -8.4095, -21.0000, -27.5954, -15.7797],
    "Longitude": [-62.2159, -38.3636, -43.2000, -48.5480, -47.9292]
})

# Juntando os dados
df_geo = df_at.merge(coordenadas, on="Regiao", how="left")
df_geo
```

```python
fig = px.scatter(df_at, x= "Ano", y= "Cobertura", color= "Regiao",  title= "Imunizações - Cobertura - Brasil")
fig.show()
```

```python
fig= px.line(df_at, x="Ano", y="Cobertura", color= "Regiao", title= "Imunizações - Cobertura - Brasil")
fig.show()
```

```python
fig= px.scatter_geo(df_geo, lat= "Latitude", lon= "Longitude", color= "Regiao", size= "Cobertura", projection= "natural earth", animation_frame= "Ano", scope= "south america", title= "Imunizações - Cobertura - Brasil")
fig.show()
```

```python
fig= px.bar (df_at, x="Ano", y="Cobertura", color= "Regiao", title= "Imunizações - Cobertura - Brasil")
fig.show()
```
